## In this notebook we show that MICE is able to fit classical copulas density and to simulate them :  indep, Gumbel and Clayton

```python
from models import mice #notre m
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
import numpy as np
from statsmodels.distributions.copula.api import (
    CopulaDistribution, GumbelCopula,ClaytonCopula,GaussianCopula,IndependenceCopula)
```

## Gumbel Copula 
Trying to learn Gumbel copula. Bivariate copula.

```python
copula = GumbelCopula(2)
sample = copula.rvs(2000) #we take 1000 samples from the copula 
```

```python
mice_gumbel = mice.MICE(dim_latent=6, dim_out=2, n_samples_train=200, n_layers=2, n_neurons=100) 
mice_gumbel.fit(sample, batch_size=200, epochs=1000)
```

We can compare our model density and the real density of a Gumbel Copula 

```python
unif_variable = np.random.uniform(0, 1, (5060, 2)) #tirage des marginales uniformes, histoire de voir... 
x,y = unif_variable[:,0],unif_variable[:,1]
gumbel_results = pd.DataFrame()
gumbel_results["x"]=x
gumbel_results["y"]=y
gumbel_results["dense"]=np.abs(np.squeeze(mice_gumbel.copulas_density.predict(unif_variable))) #densité prédite "du copule"
gumbel_results["real"] = copula.pdf(unif_variable)
gumbel_results["diff"] = np.abs(gumbel_results["dense"]-gumbel_results["real"])
```

# Clayton 

```python
copula = ClaytonCopula(2)
sample = copula.rvs(2000) #we take 1000 samples from the copula 
```

```python
mice_clayton = mice.MICE(dim_latent=6, dim_out=2, n_samples_train=200, n_layers=2, n_neurons=100) 
mice_clayton.fit(sample, batch_size=200, epochs=1000)
```

```python
unif_variable = np.random.uniform(0, 1, (5060, 2)) #tirage des marginales uniformes, histoire de voir... 
x,y = unif_variable[:,0],unif_variable[:,1]
clayton_results = pd.DataFrame()
clayton_results["x"]=x
clayton_results["y"]=y
clayton_results["dense"]=np.abs(np.squeeze(mice_clayton.copulas_density.predict(unif_variable))) #densité prédite "du copule"
clayton_results["real"] = copula.pdf(unif_variable)
clayton_results["diff"] = np.abs(clayton_results["dense"]-clayton_results["real"])
```

# Independant

```python
copula = IndependenceCopula()
sample = copula.rvs(2000) 
```

```python
mice_indep = mice.MICE(dim_latent=6, dim_out=2, n_samples_train=200, n_layers=2, n_neurons=100) 
mice_indep.fit(sample, batch_size=200, epochs=200)
```

```python
unif_variable = np.random.uniform(0, 1, (5060, 2)) #tirage des marginales uniformes, histoire de voir... 
x,y = unif_variable[:,0],unif_variable[:,1]
indep_results = pd.DataFrame()
indep_results["x"]=x
indep_results["y"]=y
indep_results["dense"]=np.abs(np.squeeze(mice_indep.copulas_density.predict(unif_variable))) #densité prédite "du copule"
indep_results["real"] = copula.pdf(unif_variable)
indep_results["diff"] = np.abs(indep_results["dense"]-indep_results["real"])
```
